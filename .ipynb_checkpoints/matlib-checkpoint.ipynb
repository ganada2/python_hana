```python
s1 = np.random.normal(loc=0,  scale=1.0, size=1000)    # loc : Mean ("centre") of the distribution.
s2 = np.random.normal(loc=5,  scale=0.5, size=1000)    # scale : Standard deviation (spread or "width") of the distribution.
s3 = np.random.normal(loc=10, scale=2.0, size=1000)    # size : Output shape
```

```python
plt.figure(figsize=(10,4))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```

```python
t = np.arange(0,5,0.01)
y = np.cos(2*np.pi*t)

plt.figure(figsize=(10,4))
plt.plot(t,y)
plt.annotate('local max', xy=(1,1), xytext=(1.0,1.2),
            arrowprops=dict(facecolor='black',shrink=0.05))
plt.ylim(-1.1,1.4)
plt.show()
```

```python
t = np.arange(0,5,0.01)
y = np.cos(2*np.pi*t)

plt.figure(figsize=(10,4))
plt.plot(t,y)
plt.annotate('local max', xy=(1,1), xytext=(1.5,2),
            arrowprops=dict(facecolor='black',shrink=0.05))
plt.ylim(-1.1,1.4)
plt.show()
```
